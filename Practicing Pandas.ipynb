````python
df.dropna() # We can use this, if we want to drop all rows havind null values.
````

````python
df.dropna(axis=1) # For dropping columns having null values.
````

````python
df.dropna(thresh=2) #Setting the threshold across the rows of 2 i.e. it will drop rows that has 2 or more null values.
````

    GROUPBY: It allows you to group together rows based off of a column and perform an aggregate function on them.

````python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
````

````python
df = pd.DataFrame(data)
````

````python
df
````

````python
# Grouping by the company name:
byComp = df.groupby('Company')
````

````python
byComp.mean() # Hence we can only do this by numeric values like sales but not by person.
````

````python
byComp.sum()
````

````python
byComp.sum().loc['FB']
````

````python
byComp.std()
````

````python
#We can directly call this groupby function after being familier with the process.
df.groupby('Company').mean().loc['GOOG']
````

````python
#Other useful functions with groupby is count()
df.groupby('Company').count()
````

````python
# For viewing all general discription, describe() function may be very useful.
df.groupby('Company').describe()
````

````python
#For transposing data to the rows side.
df.groupby('Company').describe().transpose()
````
